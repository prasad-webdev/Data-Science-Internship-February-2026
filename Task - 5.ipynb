Problem 1: Social Media - Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):

    total_likes = 0
    for like in likes_list:
        total_likes += like

    # Rule: If Total Likes >= 1000 -> Viral Post
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


analyze_post_engagement([400, 300, 350])
```

Problem 2: Healthcare - Medicine Stock Alert System

```python
def monitor_medicine_stock(stock_level):

    # Rule: Stock < 10 -> Low Stock Alert
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")


monitor_medicine_stock(6)
```

Problem 3: Agriculture - Rainfall Adequacy Checker

```python
def check_rainfall_adequacy(rainfall_data, threshold):

    total_rainfall = 0
    for value in rainfall_data:
        total_rainfall += value

    avg_rainfall = total_rainfall / len(rainfall_data)

    # Rule: Average >= required level -> Adequate
    status = "Adequate Rainfall" if avg_rainfall >= threshold else "Inadequate Rainfall"

    print(f"Average Rainfall: {avg_rainfall}")
    print(f"Rainfall Status: {status}")


check_rainfall_adequacy([70, 80, 66], 72)
```

Problem 4: Social Media - Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):

    # Rule: Use set data structure
    # If the set length is less than list length, duplicates exist
    if len(usernames) != len(set(usernames)):
        found = "Yes"
    else:
        found = "No"

    print(f"Duplicate Accounts Found: {found}")


detect_duplicate_accounts(["user_A", "user_B", "user_A"])
```

Problem 6: Agriculture - Premium Crop Price Filter

```python
def filter_premium_crops(price_list):

    # Rule: Price > ₹2000 -> Premium
    premium_crops = [price for price in price_list if price > 2000]

    print(f"Premium Crops: {premium_crops}")

filter_premium_crops([1500, 2500, 3200, 1800])
```

Problem 7: System Monitoring - Application Health Checker

```python
def evaluate_app_health(error_count):

    # Rules: 0 = Healthy, <=5 = Minor, >5 = Critical
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")


evaluate_app_health(7)
```

Problem 8: Banking - Daily Transaction Limit Checker

```python
def check_transaction_limit(amount):

    # Rule: Limit = 50,000
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")


check_transaction_limit(60000)
```

Problem 9: E-Learning - Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_record):

    days_present = sum(attendance_record)
    total_days = len(attendance_record)
    percentage = (days_present / total_days) * 100

    # Rule: Attendance >= 75% -> Eligible
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

check_exam_eligibility([1, 1, 1, 1, 0])
```

Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_smart_bill(units):

    bill = 0
    # Billing Rules: Slabs of 3, 5, and 7
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7


    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


total_bill, level = calculate_smart_bill(250)
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {level}")
```
