Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name

    def set_status(self, status):
        print(f"{self.name} is {status}")

light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

light = SmartLight(light_name)
light.set_status(action)
```

Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("\nEmployee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display_card()
```

Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

c_name = input("Contact Name: ")
c_phone = input("Phone Number: ")

contact = Contact(c_name, c_phone)
contact.show_contact()
```

Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        print(f"\nProduct: {self.name}")
        print(f"Price: ₹{self.price}")

p_name = input("Product Name: ")
p_price = input("Price: ")

product = Product(p_name, p_price)
product.print_tag()
```

Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"\nMovie: {self.name}")
        print(f"Rating: {self.rating}/5")

m_name = input("Movie Name: ")
m_rating = input("Rating: ")

movie = Movie(m_name, m_rating)
movie.display_details()
```

Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_delivery(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer}")
        print(f"Address: {self.address}")

cust_name = input("Customer Name: ")
addr = input("Address: ")

delivery = Delivery(cust_name, addr)
delivery.show_delivery()
```
