Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        try:
            amount = int(input("Enter recharge amount: "))
            # Amount must be >= 50 and in the valid plans list
            if amount >= 50 and amount in valid_plans:
                print(f"Recharge of ₹{amount} successful!")
                break
            else:
                print("Invalid amount or plan. Please try again.")
        except ValueError:
            print("Please enter a valid numeric value.")

validate_recharge()
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert (Stock: {quantity})")
        else:
            print(f"{product}: Stock OK")

# Create sample data and CALL the function
my_stock = {"Apples": 20, "Bananas": 10, "Oranges": 15}
check_inventory(my_stock)
```

Problem 3: Student Result Processing System

```python
def process_result(marks_list):
    total = 0
    for mark in marks_list:
        total += mark

    average = total / len(marks_list)

    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"


student_marks = list(map(int, input("Enter marks separated by space: ").split()))
print(process_result(student_marks))
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(km, is_peak):
    base_fare = 50

    fare = base_fare + (km * 12)

    if is_peak.lower() == 'yes':
        fare += fare * 0.25

    return fare

while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)
        print(f"Estimated Fare: ₹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (y/n): ")
        if retry.lower() != 'y':
            break
    except ValueError:
        print("Invalid input. Please enter numbers for distance.")
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance_list):
    present_count = 0
    for status in attendance_list:
        if status == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = input("Enter attendance (P/A without space): ").upper()
print(f"Status: {check_eligibility(attendance)}")
```

Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak: Minimum 8 characters required."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must contain a digit and a special character (@, #, or $)."


pwd = input("Enter password to check: ")
print(check_password_strength(pwd))
```
